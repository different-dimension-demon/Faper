```python
import numpy as np
import matplotlib.pyplot as plt

# Input data
model_names = ['SpikingCard']
data = np.array([
    [6.75, 16.02, 30.15], 
    [2.66, 4.47, 13.92], 
    [2.42, 4.69, 16.78], 
    [19.32, 105.69, 362.32], 
    [10.34, 50.90, 500.92], 
    [3.78, 14.25, 37.88]  
])

# Define the thresholds and metrics
train_thresholds = ['4-Join', '6-Join', '8-Join']
test_thresholds = ['6-Join', '8-Join']
metrics = ['50th', '75th', '90th']

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(20, 8))

# Colors for bar and line plots
colors = ['#D8BFD8', '#ADD8E6', '#FFA07A', '#FFB6C1']
line_colors = ['#800080', '#00BFFF', '#FF8C00', '#FF1493']

# Plot for test set '6-Join'
x = np.arange(len(train_thresholds))
width = 0.15
for i, metric in enumerate(metrics):
    axs[0].bar(x + i * width, data[:3, i], width, color=colors[i], label=metric)
    axs[0].plot(x + i * width, data[:3, i], marker='o', color=line_colors[i], linewidth=2, label=f"{metric} Line")

axs[0].set_title('Training on Different Thresholds, Testing on 6-Join', fontsize=16)
axs[0].set_xlabel('Training Data Set', fontsize=14)
axs[0].set_ylabel('Estimation q_error', fontsize=14)
axs[0].set_xticks(x + 1.5 * width)
axs[0].set_xticklabels(train_thresholds, fontsize=12)
axs[0].legend(fontsize=12)

# Plot for test set '8-Join'
for i, metric in enumerate(metrics):
    axs[1].bar(x + i * width, data[3:, i], width, color=colors[i], label=metric)
    axs[1].plot(x + i * width, data[3:, i], marker='o', color=line_colors[i], linewidth=2, label=f"{metric} Line")

axs[1].set_title('Training on Different Thresholds, Testing on 8-Join', fontsize=16)
axs[1].set_xlabel('Training Threshold', fontsize=14)
axs[1].set_ylabel('Estimation q_error', fontsize=14)
axs[1].set_xticks(x + 1.5 * width)
axs[1].set_xticklabels(train_thresholds, fontsize=12)
axs[1].legend(fontsize=12)

# Save the figure as a PDF
plt.savefig('output.pdf', format='pdf', dpi=600)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 输入数据
model_names = ['SpikingCard']
data = np.array([
    [2.91, 8.01, 30.26, 30.81],  # No filtering
    [2.86, 7.29, 23.90, 27.62],  # Threshold 1.0
    [2.86, 7.41, 25.08, 25.64],  # Threshold 0.8
    [2.62, 6.94, 24.57, 25.21],  # Threshold 0.7
    [2.44, 6.03, 16.11, 15.96],  # Threshold 0.5
    [2.45, 6.08, 19.65, 12.61]   # Threshold 0.3
])

# 定义阈值和指标
thresholds = ['No filtering', '1.0', '0.8', '0.7', '0.5', '0.3']
metrics = ['50th', '75th', '90th', 'Mean']

# 创建一个固定大小的图形
fig, ax = plt.subplots(figsize=(6, 4.5))

# 绘制柱状图
x = np.arange(len(thresholds))
width = 0.15
colors = ['#5861AC', '#F28080', '#A5C2E2', '#6AB4C1', '#72C3A3']

for i, metric in enumerate(metrics):
    ax.bar(x + i * width, data[:, i], width, color=colors[i], edgecolor='black', linewidth=1, label=metric)

# 设置标题和坐标轴标签
ax.set_title('Comparison Across Different Thresholds of Uncertainty', fontsize=13, fontweight='bold')
ax.set_xlabel('Threshold', fontsize=13)
ax.set_ylabel('Estimation q_error', fontsize=13)

# 设置x轴刻度和标签
ax.set_xticks(x + 1.5 * width)
ax.set_xticklabels(thresholds, fontsize=11)

# 添加图例
ax.legend(fontsize=11)

# 确保布局紧凑
fig.tight_layout()

# 保存为 PDF 格式
plt.savefig('output.pdf', format='pdf', dpi=600)

# 显示图像
plt.show()
```
